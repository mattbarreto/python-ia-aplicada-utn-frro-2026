## ◈ Microglosario

| Término | Qué es en lenguaje llano |
|---|---|
| **Repositorio** | La carpeta del proyecto versionada por Git. Contiene todos los archivos y el historial completo de cambios. |
| **Commit** | Un punto de control guardado: una fotografía del estado de los archivos en un momento determinado, con un mensaje que explica qué cambió. |
| **Staging area** | La sala de espera antes del commit. Ahí se colocan los archivos que van a formar parte del siguiente punto de control. |
| **Rama (branch)** | Una línea de desarrollo paralela dentro del repositorio. La rama principal se llama `main`. |
| **Remoto (remote)** | La copia del repositorio que vive en la nube, en GitHub. Se sincroniza con la copia local mediante `push` y `pull`. |
| **Clonar** | Descargar por primera vez un repositorio remoto a tu computadora, incluyendo todo su historial. |

## 🗺️ Parte 1: Qué son Git y GitHub

### Analogía

Git es el sistema de guardado inteligente de un videojuego: no solo guarda dónde estás ahora, sino cada checkpoint que marcaste en el camino. Podés volver a cualquier punto anterior sin perder lo que construiste después. GitHub es el servidor en la nube donde ese historial queda publicado y accesible desde cualquier lugar.

### Dónde vive esto en el mundo real

Git es el estándar de la industria del software. Todo equipo de desarrollo — desde una startup de dos personas hasta ingenieras y ingenieros en Google, Microsoft o Mercado Libre — versiona su código con Git. En el ámbito científico y de datos, los repositorios de modelos, datasets y papers reproducibles también se publican en GitHub. En este curso, tu repositorio personal es tu portafolio de aprendizaje: documenta lo que construiste encuentro a encuentro.

---

**Git** es un sistema de control de versiones. Te permite registrar cambios en tus archivos, volver a estados anteriores y trabajar de manera ordenada.

**GitHub** es una plataforma que aloja repositorios Git en la web. Funciona como espacio de publicación, respaldo y entrega.

En este curso, tu repositorio personal cumple varias funciones al mismo tiempo:

- organiza tus materiales encuentro a encuentro
- conserva notebooks y archivos de cada clase
- permite a los docentes revisar tu trabajo
- simplifica las entregas, porque compartís enlaces ya publicados

Tu repositorio no es un depósito improvisado de archivos. Es tu carpeta de cursada versionada.

## Parte 2: Convención para nombrar el repositorio

El nombre del repositorio debe respetar esta estructura:

```text
apellido-nombre-materia-cuatrimestre-anio
```

Ejemplos correctos:

- `perez-ana-python-1c-2026`
- `gomez-juan-ia-2c-2027`
- `lopez-maria-python-1c-2026`

Componentes del nombre:

| Parte | Significado | Ejemplo |
|---|---|---|
| `apellido` | tu apellido | `perez` |
| `nombre` | tu nombre | `ana` |
| `materia` | nombre corto del curso | `python` |
| `1c` o `2c` | período o cuatrimestre | `1c` |
| `año` | año lectivo | `2026` |

Reglas prácticas:

- usá **minúsculas** en todo el nombre
- usá **guiones medios** (`-`) en lugar de espacios
- no uses tildes ni `ñ` — reemplazala por `n` simple
- mantené siempre el mismo criterio de escritura

## Parte 3: Estructura interna del repositorio

Dentro del repositorio, la organización sigue un orden temático por encuentro.

Formato del nombre de carpeta:

```text
NNN_tema
```

Donde `NNN` es un número de tres dígitos correlativos y `tema` es una descripción corta en minúsculas, separada por guiones bajos.

| Carpeta | Encuentro |
|---|---|
| `001_entorno` | Encuentro 1: entorno de desarrollo, Git y GitHub |
| `002_python` | Encuentro 2: Python aplicado |
| `003_docker` | Encuentro 3: Docker y reproducibilidad |

Ejemplo de estructura completa:

```text
perez-ana-python-1c-2026/
├── 001_entorno/
│   ├── 001_git.ipynb
│   └── 002_python.ipynb
├── 002_python/
│   ├── 001_lenguaje.ipynb
│   ├── 002_memoria.ipynb
│   └── 003_servicios.ipynb
└── README.md
```

No es necesario definir desde ahora cuántas carpetas tendrá el repositorio. Lo importante es mantener el patrón de tres dígitos a medida que avanzás en los encuentros.

### Regla clave: carpetas vacías

Git **no versiona carpetas vacías**. Si creás una carpeta sin archivos adentro, no va a aparecer en GitHub.

Dos caminos válidos:

1. Crear las carpetas a medida que tenés material para guardar.
2. Dejar la estructura preparada con un archivo de apoyo dentro:

```bash
mkdir -p 001_entorno
touch 001_entorno/.gitkeep
```

`.gitkeep` es un archivo vacío convencional cuyo único propósito es que Git detecte la carpeta.

## Parte 4: Preparación del entorno

Antes de crear el repositorio, preparar estas tres cosas:

1. **Instalar Git** desde [git-scm.com](https://git-scm.com/downloads)
2. **Crear cuenta en GitHub** desde [github.com/signup](https://github.com/signup)
3. **Elegir una ubicación local de trabajo** en tu computadora — por ejemplo, una carpeta llamada `Cursada` o `Repositorios`

Recomendaciones:

- en Windows, instalá Git con **Git Bash** y **Git Credential Manager**
- usá el mismo correo en Git y en GitHub
- configurá el repositorio como **público** — así los enlaces de entrega funcionan sin pedir permisos al docente

Para verificar que Git quedó instalado:

```bash
git --version
```

### Configuración de identidad en Git

Git necesita saber quién registra los cambios. Configurarlo una sola vez:

```bash
git config --global user.name "Tu Nombre y Apellido"
git config --global user.email "tu-correo@ejemplo.com"
git config --global init.defaultBranch main
```

Para revisar la configuración actual:

```bash
git config --global --list
```

Si Git muestra el error `Author identity unknown`, casi siempre se resuelve con estos comandos.

## Parte 5: Crear el repositorio en GitHub

La ruta recomendada: crear primero en GitHub y después clonar a la computadora.

Pasos:

1. Entrá a GitHub
2. Hacé clic en **New repository**
3. Escribí el nombre con la convención acordada
4. Elegí visibilidad **Public**
5. Marcá **Add a README file** para que la rama `main` quede creada
6. Creá el repositorio

Revisá el nombre dos veces antes de confirmar. Corregirlo al principio es fácil; corregirlo tarde genera trabajo innecesario.

## Parte 6: Clonar el repositorio

Una vez creado el repositorio remoto, copiamos su contenido a la máquina local con `git clone`.

En GitHub:

1. Abrí el repositorio
2. Hacé clic en **Code**
3. Copiá la URL HTTPS

En la terminal, ubicarse en la carpeta de trabajo y ejecutar:

```bash
git clone https://github.com/tu-usuario/perez-ana-python-1c-2026.git
cd perez-ana-python-1c-2026
```

A partir de ese momento, todo lo que guardemos dentro de esa carpeta puede versionarse con Git.

## ⚡ Parte 8: Primer flujo completo de Git

El ciclo fundamental de Git son cuatro comandos. Ejecutarlos dentro del repositorio:

```bash
git status
git add .
git commit -m "Crear estructura inicial del repositorio"
git push origin main
```

Qué hace cada paso:

| Comando | Qué hace |
|---|---|
| `git status` | Muestra qué archivos cambiaron y cuáles están en el staging area |
| `git add .` | Mueve todos los cambios al staging area (la sala de espera del commit) |
| `git commit -m "..."` | Guarda un punto de control local con un mensaje descriptivo |
| `git push origin main` | Publica ese punto de control en GitHub |

Cuando termine, actualizar la página del repositorio en GitHub y verificar que la estructura de carpetas aparezca publicada.

- ¿Qué diferencia hay entre `git add .` y `git add nombre_archivo`? ¿Cuándo conviene usar uno u otro?
- ¿Qué pasa si hacemos `git push` sin haber hecho `git commit` antes?

## Parte 9: Flujo cotidiano durante la cursada

A partir del primer push, el flujo habitual antes y después de cada clase:

```bash
# Al inicio de la sesión: traer cambios del repositorio remoto
git pull origin main

# Trabajar, editar o agregar archivos...

# Al cerrar la sesión: registrar y publicar el trabajo
git status
git add .
git commit -m "Mensaje descriptivo"
git push origin main
```

Ejemplos de mensajes de commit útiles:

- `enc-01: agregar notebook de git y primer flujo`
- `enc-02: script integrador con api de clima`
- `enc-03: dockerfile del entorno de trabajo`

Regla de trabajo: siempre empezar la sesión con `git pull`. Siempre terminarla con `git push`.

## Parte 10: Subir notebooks, imágenes, PDF y diapositivas

El repositorio no guarda solo código. Funciona como carpeta ordenada de materiales de cursada.

| Tipo de archivo | Extensiones | Recomendaciones |
|---|---|---|
| Notebooks | `.ipynb` | Guardarlos dentro de la carpeta del encuentro correspondiente |
| Scripts | `.py` | Mismo criterio que los notebooks |
| Imágenes | `.png`, `.jpg` | Dentro de la carpeta del encuentro, no sueltas en la raíz |
| Documentos | `.pdf` | Se previsualiza directamente en GitHub |
| Diapositivas | `.pptx`, `.pdf` | Preferir PDF para compatibilidad de visualización |

El flujo para subir cualquier tipo de archivo es siempre el mismo: `add` → `commit` → `push`.

## Parte 11: Entregas por enlace

Cuando los docentes pidan enviar un material por formulario, la entrega recomendada es el **enlace del archivo publicado en GitHub**.

Ventajas:

- no hay que comprimir archivos
- no depende de permisos de Google Colab
- el material queda respaldado y versionado
- los docentes ven exactamente la versión publicada en la fecha de entrega

Procedimiento:

1. Hacer `push` del archivo
2. Abrirlo en GitHub y verificar que cargue correctamente
3. Copiar la URL completa desde la barra del navegador
4. Pegar ese enlace en el formulario de entrega

## Parte 12: Ramas y pull requests — versión simplificada

En un repositorio personal de cursada, casi siempre se trabaja directamente sobre `main`. Eso es suficiente para el flujo básico.

Si los docentes piden trabajar con ramas — para correcciones aisladas o trabajo grupal —, el esquema básico es:

```bash
git switch -c correccion-enc-002
# realizar cambios
git add .
git commit -m "Corregir ejercicio del encuentro 002"
git push -u origin correccion-enc-002
```

Dos ideas clave para recordar:

- `main` es la rama principal de entrega y trabajo
- una rama alternativa sirve para aislar cambios antes de integrarlos a `main`

## 🧭 Parte 13: Troubleshooting

### 1. `git` no se reconoce en la terminal

**Causa:** Git no se instaló correctamente o la terminal no se reinició.  
**Solución:** reinstalar Git y volver a abrir la terminal. Verificar con `git --version`.

---

### 2. `Author identity unknown`

**Causa:** Git no tiene configurado nombre ni correo.  
**Solución:**
```bash
git config --global user.name "Tu Nombre y Apellido"
git config --global user.email "tu-correo@ejemplo.com"
```

---

### 3. `fatal: Authentication failed`

**Causa:** la autenticación con GitHub falló.  
**Solución:** usar un **Personal Access Token (PAT)** o autenticación SSH cuando GitHub lo pida. No usar la contraseña normal de la cuenta.

---

### 4. `! [rejected] main -> main (non-fast-forward)`

**Causa:** el repositorio remoto tiene cambios que la copia local no incorporó.  
**Solución:**
```bash
git pull origin main
git push origin main
```

---

### 5. La carpeta no aparece en GitHub

**Causa:** la carpeta estaba vacía o faltaron pasos del flujo.  
**Solución:** agregar un archivo dentro (o un `.gitkeep`) y ejecutar `add` → `commit` → `push`.

---

### 6. El archivo existe localmente pero no se ve en GitHub

**Causa:** faltó algún paso entre `add`, `commit` y `push`.  
**Solución:**
```bash
git status
git add .
git commit -m "Subir archivo faltante"
git push origin main
```

---

### 7. El enlace pide permisos

**Causa:** se compartió el enlace de Colab en lugar del de GitHub, o el repositorio es privado.  
**Solución:** compartir siempre el enlace del archivo publicado en GitHub. Configurar el repositorio como público.

---

### 8. El repositorio tiene un nombre incorrecto

**Causa:** no se siguió la convención de nomenclatura al crearlo.  
**Solución:** si recién se creó, conviene empezar de cero. Si ya tiene contenido, renombrarlo desde los ajustes en la web de GitHub y actualizar la URL remota en la terminal local.
