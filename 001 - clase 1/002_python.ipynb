## ⚙️ Variables y tipos de datos

### Analogía

Una variable es una etiqueta pegada a una caja. El nombre de la etiqueta es el nombre de la variable; el contenido de la caja es el valor. Python mira la caja y decide qué tipo de contenedor usar según lo que hay adentro.

### Dónde vive esto en el mundo real

Todo sistema de software guarda información en variables. Una app de turnos médicos tiene variables para el nombre del paciente, la fecha del turno, el número de consultorio y si el turno está confirmado. Cada una de esas variables tiene un tipo que determina cómo el sistema la almacena, valida y muestra.

```python
# Los cuatro tipos básicos de Python
nombre_ciudad  = "Rosario"    # str  — texto
poblacion      = 1_700_000    # int  — número entero
superficie_km2 = 178.69       # float — número decimal
es_capital     = False        # bool — verdadero / falso

print("✓ Variables creadas.")
```

```python
# type() revela el tipo de cada variable
print(f"nombre_ciudad  → {type(nombre_ciudad)}")
print(f"poblacion      → {type(poblacion)}")
print(f"superficie_km2 → {type(superficie_km2)}")
print(f"es_capital     → {type(es_capital)}")
```

```python
# Operaciones aritméticas básicas
valor_a = 5
valor_b = 10

suma           = valor_a + valor_b
resta          = valor_a - valor_b
multiplicacion = valor_a * valor_b
division       = valor_a / valor_b       # siempre devuelve float
division_entera = valor_a // valor_b     # descarta los decimales
resto          = valor_a % valor_b       # el resto de la división
potencia       = valor_a ** 2            # valor_a elevado al cuadrado

print(f"Suma           : {suma}")
print(f"Resta          : {resta}")
print(f"Multiplicación : {multiplicacion}")
print(f"División       : {division}")
print(f"División entera: {division_entera}")
print(f"Resto          : {resto}")
print(f"Potencia       : {potencia}")
```

```python
# Los operadores de comparación devuelven bool (True o False)
print(f"¿valor_a == valor_b? → {valor_a == valor_b}")
print(f"¿valor_a < valor_b?  → {valor_a < valor_b}")
print(f"¿valor_a >= valor_b? → {valor_a >= valor_b}")
```

## 🧬 Cadenas de texto (strings)

### Analogía

Un string es como una hilera de perlas: cada carácter es una perla en una posición fija. Podés acceder a cualquier perla por su número de posición, cortar la hilera donde quieras, y aplicar transformaciones que producen una hilera nueva.

### Dónde vive esto en el mundo real

El procesamiento de texto es una de las tareas más comunes en software. Los chatbots limpian y normalizan el texto antes de procesarlo. Los sistemas de búsqueda aplican `.lower()` a las consultas para que "Python" y "python" devuelvan los mismos resultados. En los encuentros 6 y 8 vamos a usar LLMs que reciben y producen strings como entrada y salida principal.

```python
frase = "programación en Python"

# Acceso por índice — el primer carácter está en la posición 0
primer_caracter = frase[0]
ultimo_caracter = frase[-1]    # -1 accede desde el final

print(f"Primer carácter: {primer_caracter}")
print(f"Último carácter: {ultimo_caracter}")
```

```python
# Slicing: extraer una porción del string [inicio:fin:paso]
primeras_tres   = frase[:3]       # desde el inicio hasta la posición 3 (sin incluir)
del_cinco       = frase[5:]       # desde la posición 5 hasta el final
al_reves        = frase[::-1]     # paso -1 invierte la cadena

print(f"Primeras tres letras : {primeras_tres}")
print(f"Desde posición 5     : {del_cinco}")
print(f"Al revés             : {al_reves}")
```

```python
# Métodos útiles de strings — todos devuelven un string nuevo, no modifican el original
en_mayusculas   = frase.upper()
en_titulo       = frase.title()
palabras        = frase.split()         # divide en una lista usando el espacio como separador
sin_espacios    = "   texto con espacios   ".strip()
reemplazado     = frase.replace("Python", "Argentina")

print(f"upper()   : {en_mayusculas}")
print(f"title()   : {en_titulo}")
print(f"split()   : {palabras}")
print(f"strip()   : '{sin_espacios}'")
print(f"replace() : {reemplazado}")
```

```python
# f-strings: la forma moderna de combinar variables y texto
ciudad     = "Rosario"
habitantes = 1_700_000

descripcion = f"{ciudad} tiene aproximadamente {habitantes:,} habitantes."
print(descripcion)
```

## 🗂️ Listas

### Analogía

Una lista es como una fila numerada de casilleros: cada casillero tiene un número (el índice) y puede guardar cualquier cosa. Podés agregar casilleros al final, quitar uno del medio, o acceder directamente al número 7 sin recorrer los anteriores.

### Dónde vive esto en el mundo real

Las listas son la estructura de datos más usada en Python. Los resultados de una búsqueda son una lista. Los tweets de un timeline son una lista. Los registros de una base de datos que procesamos en el Encuentro 5 con Pandas son, en su forma más básica, listas de listas. Todo lo que ocurre en volumen pasa por una lista.

```python
# Lista de participantes del encuentro
participantes = ["Alicia", "Bruno", "Camila", "Diego", "Ernestina"]

# Lista de edades en el mismo orden
edades = [32, 45, 28, 37, 51]

print(f"✓ {len(participantes)} participantes registrados.")
```

```python
# Acceso por índice
primera = participantes[0]     # primer elemento (índice 0)
ultima  = participantes[-1]    # último elemento

print(f"Primera: {primera}")
print(f"Última : {ultima}")
```

```python
# Métodos principales
participantes.append("Federico")          # agrega al final
participantes.remove("Bruno")             # elimina por valor
participantes.insert(1, "Beatriz")        # inserta en la posición indicada

print(f"Lista actualizada: {participantes}")
```

```python
# Operaciones sobre listas numéricas
print(f"Total de edades   : {sum(edades)}")
print(f"Promedio de edades: {sum(edades) / len(edades):.1f}")
print(f"Edad mínima       : {min(edades)}")
print(f"Edad máxima       : {max(edades)}")
print(f"Ordenadas         : {sorted(edades)}")
```

## 🔑 Diccionarios

### Analogía

Un diccionario es como la ficha de una persona en un registro civil: cada dato tiene una etiqueta (la clave) y un valor asociado. En lugar de buscar por número de posición como en una lista, buscamos por nombre de campo. `"nombre"` siempre encuentra el nombre, sin importar cuántos otros campos haya.

### Dónde vive esto en el mundo real

Los diccionarios son la estructura natural del JSON — el formato de intercambio de datos de toda la web. Cuando una API devuelve información sobre el clima, sobre un usuario, o sobre un pago, esa información llega como un diccionario (o como una lista de diccionarios). En el Encuentro 2 ya vamos a trabajar con JSON directamente.

```python
# Ficha de una participante del curso
ficha_participante = {
    "nombre": "Alicia Fernández",
    "edad": 32,
    "ciudad": "Rosario",
    "experiencia_anios": 3,
    "activa": True
}

print("✓ Diccionario creado.")
```

```python
# Acceso por clave
nombre_participante = ficha_participante["nombre"]
edad_participante   = ficha_participante["edad"]

print(f"Nombre : {nombre_participante}")
print(f"Edad   : {edad_participante}")
```

```python
# Agregar, modificar y eliminar campos
ficha_participante["email"] = "alicia@ejemplo.com"    # agregar campo nuevo
ficha_participante["edad"]  = 33                       # modificar campo existente
del ficha_participante["activa"]                       # eliminar campo

print(ficha_participante)
```

```python
# Recorrer todas las claves y valores
for clave in ficha_participante.keys():
    valor = ficha_participante[clave]
    print(f"  {clave:20} → {valor}")
```

## 🧭 Condicionales

### Analogía

Un condicional es como un semáforo: evalúa una condición y elige un camino. Verde → seguís. Rojo → parás. Amarillo → algo en el medio. El programa no puede tomar dos caminos al mismo tiempo — siempre elige uno.

### Dónde vive esto en el mundo real

Toda lógica de negocio es una cadena de condicionales. Un sistema bancario evalúa si el saldo es suficiente antes de aprobar una transferencia. Un algoritmo de moderación de contenido clasifica publicaciones según criterios definidos con `if`. En los agentes de IA del Encuentro 9, los condicionales determinan qué herramienta ejecutar según el pedido del usuario.

```python
# Condicional simple: un número secreto para adivinar
numero_secreto  = 42
intento_usuario = 35

if intento_usuario == numero_secreto:
    print("✓ ¡Correcto! Adivinaste el número.")
elif intento_usuario < numero_secreto:
    print("▲ El número secreto es mayor.")
else:
    print("▼ El número secreto es menor.")
```

```python
# Condicionales con operadores lógicos
temperatura = 38.5
humedad     = 85

if temperatura > 35 and humedad > 80:
    condicion_clima = "Calor extremo con alta humedad — riesgo para trabajo a la intemperie"
elif temperatura > 35 or humedad > 80:
    condicion_clima = "Condiciones adversas — precaución recomendada"
else:
    condicion_clima = "Condiciones normales"

print(f"Temperatura: {temperatura}°C | Humedad: {humedad}%")
print(f"Condición  : {condicion_clima}")
```

## ⚡ Bucles

### Analogía

Un bucle es como un empleado que sigue la misma instrucción para cada ítem de una lista: abre el cajón 1, revisa, cierra. Abre el cajón 2, revisa, cierra. Y así hasta terminar. La instrucción es siempre la misma — lo que cambia es el ítem.

### Dónde vive esto en el mundo real

Los bucles procesan volumen. Un sistema de facturación recorre miles de líneas de una planilla. Un modelo de machine learning itera sobre millones de ejemplos de entrenamiento. Un bot de notificaciones recorre la lista de usuarios y envía un mensaje a cada uno. En el Encuentro 2 vamos a usar bucles para procesar todos los registros de una API y generar reportes automáticamente.

```python
# Bucle for — recorre cada elemento de la lista
nombres = ["Alicia", "Bruno", "Camila", "Diego", "Ernestina"]

for nombre_actual in nombres:
    print(f"Hola, {nombre_actual}.")
```

```python
# Bucle for con range() — útil para repetir N veces o usar el índice
print("Primeros 5 múltiplos de 3:")

for numero in range(1, 6):    # genera 1, 2, 3, 4, 5
    multiplo = numero * 3
    print(f"  3 × {numero} = {multiplo}")
```

```python
# Bucle for con índice — para recorrer dos listas en paralelo
nombres_lista = ["Alicia", "Bruno", "Camila", "Diego"]
notas_lista   = [8.5, 6.0, 9.5, 4.0]

for indice in range(len(nombres_lista)):
    nombre_actual = nombres_lista[indice]
    nota_actual   = notas_lista[indice]

    if nota_actual >= 6:
        estado = "Aprobada/o"
    else:
        estado = "Desaprobada/o"

    print(f"  {nombre_actual:10} | Nota: {nota_actual} | {estado}")
```

```python
# Bucle while — repite mientras una condición sea verdadera
contador = 0
limite   = 5

while contador < limite:
    contador = contador + 1    # incrementamos el contador en cada vuelta
    print(f"  Vuelta {contador} de {limite}")

print("✓ Bucle completado.")
```

## 🔧 Funciones

### Analogía

Una función es una máquina con nombre: le ponés ingredientes (parámetros), la máquina hace algo, y puede darte un resultado (return). Una vez que construiste la máquina, la usás tantas veces como querés sin volver a construirla.

### Dónde vive esto en el mundo real

Las funciones son el bloque fundamental de todo software. Un endpoint de una API es una función que el servidor expone al mundo. Los modelos de machine learning son composiciones de miles de funciones matemáticas. En este curso, las primeras funciones que escribamos hoy van a crecer hasta convertirse en los endpoints del Encuentro 4 y los agentes del Encuentro 9.

```python
# Definimos una función — todavía no hace nada, solo existe
def saludar(nombre_persona):
    saludo = f"Hola, {nombre_persona}. Bienvenida/o al curso."
    return saludo

print("✓ Función definida.")
```

```python
# Llamamos a la función — ahora sí se ejecuta
mensaje = saludar("Alicia")
print(mensaje)

# La misma función, distinto argumento
print(saludar("Bruno"))
```

```python
# Función con múltiples parámetros y lógica interna
def mostrar_en_mayusculas(texto, repeticiones):
    texto_en_mayusculas = texto.upper()

    for vez in range(repeticiones):
        print(f"  [{vez + 1}] {texto_en_mayusculas}")

mostrar_en_mayusculas("Python es genial", 3)
```

```python
# Las funciones pueden llamarse dentro de un bucle
participantes_lista = ["Alicia", "Bruno", "Camila"]

for nombre_participante in participantes_lista:
    mensaje_bienvenida = saludar(nombre_participante)    # llamamos a la función definida antes
    print(mensaje_bienvenida)
```
