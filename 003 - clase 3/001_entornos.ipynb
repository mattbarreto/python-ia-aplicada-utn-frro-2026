## 🧭 El problema que vamos a resolver

### Analogía

Imaginen que comparten un taller con otras personas. Cada una trabaja en proyectos distintos: una necesita un martillo de 200g, otra uno de 500g. Si hay un solo martillo compartido en el taller, hay conflicto. La solución no es comprar un martillo universal que haga todo a la vez —eso no existe—, sino tener cajones separados, uno por proyecto, donde cada persona guarda exactamente lo que necesita.

Los entornos de desarrollo funcionan igual. Cada proyecto necesita versiones específicas de Python y sus librerías. Sin aislamiento, los proyectos se pisan entre sí.

### Dónde vive esto en el mundo real

El problema de los entornos es cotidiano en el trabajo profesional. En una empresa de software, cada equipo tiene su propio stack de versiones. Los sistemas de integración continua (CI/CD) crean entornos limpios para cada compilación. Los servicios en la nube corren en contenedores precisamente para garantizar que lo que funcionó en desarrollo también funcione en producción. En este curso, los entornos aislados van a aparecer en cada encuentro: `uv` hoy, Docker desde el bloque siguiente.

---

### El conflicto de versiones

En el desarrollo de software, es habitual trabajar con proyectos que tienen requisitos distintos:

| Proyecto | Python | Librería clave |
|---|---|---|
| Proyecto A | 3.9 | Django 3.2 |
| Proyecto B | 3.12 | Django 5.0 |
| Proyecto C | 3.11 | FastAPI + PostgreSQL |

Si instalamos todo directamente en el sistema operativo, surgen cuatro problemas:

1. **Conflictos de versiones** — dos proyectos necesitan versiones diferentes de la misma librería
2. **Contaminación del sistema** — cientos de paquetes globales que nadie sabe bien qué hacen
3. **Dificultad para compartir** — otros no pueden replicar el entorno
4. **Limpieza imposible** — no sabemos qué desinstalar sin romper algo

## ⚙️ Entornos virtuales de Python

### Analogía

Un entorno virtual es el cajón individual del taller. Cada proyecto tiene el suyo. Lo que está en el cajón de un proyecto no interfiere con los demás. Si el cajón se rompe, lo descartamos y creamos uno nuevo sin afectar nada más.

### Dónde vive esto en el mundo real

Todos los proyectos Python profesionales usan entornos virtuales. Es el estándar de facto desde Python 3.3. Las herramientas modernas como `uv` (que usamos en este curso) los gestionan en segundos. Cuando clonamos un repositorio de GitHub, el README siempre incluye cómo crear y activar el entorno virtual antes de instalar dependencias.

---

Un entorno virtual es un **directorio autocontenido** que incluye:

- Una instalación de Python específica
- Un conjunto de paquetes instalados
- Scripts de activación

### Herramientas disponibles

| Herramienta | Cuándo usarla |
|---|---|
| `uv venv` | **Este curso** — ultrarrápido, moderno, reemplaza pip+venv |
| `python -m venv` | Incluido con Python, sin instalación extra |
| `conda` | Ciencia de datos con dependencias C/C++ |

### Con `uv` — el flujo de este curso

```bash
# Crear el entorno virtual:
uv venv .venv

# Activarlo (Linux/macOS):
source .venv/bin/activate

# Activarlo (Windows PowerShell):
.venv\Scripts\Activate.ps1

# Instalar dependencias:
uv pip install -r requirements.txt
```

### Ventajas y limitaciones

| Aspecto | Detalle |
|---|---|
| ✓ Livianos | Solo ocupan espacio para paquetes de Python |
| ✓ Rápidos | Se crean en segundos con `uv` |
| ✓ Integración nativa | VS Code, JupyterLab los detectan automáticamente |
| ✗ Solo Python | No aíslan servicios externos (bases de datos, APIs) |
| ✗ Portabilidad limitada | Puede haber diferencias entre sistemas operativos |

## 🗂️ Contenedores Docker

### Analogía

Si el entorno virtual es un cajón individual, Docker es un taller entero dentro de una caja. La caja incluye las herramientas, la mesa, la iluminación y hasta el piso. Se puede enviar esa caja completa a cualquier parte del mundo y el taller funciona igual: misma iluminación, misma distribución, mismas herramientas.

### Dónde vive esto en el mundo real

Todos los sistemas a escala usan contenedores. Netflix, Spotify, Mercado Libre despliegan miles de contenedores en paralelo. Los servicios de machine learning en producción —desde APIs de modelos hasta pipelines de entrenamiento— corren en contenedores porque garantizan reproducibilidad y escalabilidad. En el próximo bloque vamos a construir nuestra propia imagen Docker.

---

Un contenedor empaqueta la aplicación con **todo lo que necesita**:

- Sistema operativo base
- Lenguaje de programación y versión
- Dependencias y librerías
- Código de la aplicación

### Ventajas y limitaciones

| Aspecto | Detalle |
|---|---|
| ✓ Aislamiento completo | Sistema operativo, servicios, configuraciones |
| ✓ Portabilidad total | Funciona igual en cualquier máquina con Docker |
| ✓ Múltiples servicios | Base de datos, cache, API en contenedores separados |
| ✓ Estándar de producción | El mismo contenedor que se desarrolla se despliega |
| ✗ Curva de aprendizaje | Conceptos nuevos: imagen, capa, registro, orquestación |
| ✗ Overhead de recursos | Más RAM y disco que un entorno virtual |

### ✎ Para pensar

## ◈ Comparación lado a lado

| Característica | Entorno virtual (`uv`) | Docker |
|---|---|---|
| Aislamiento | Solo Python | Sistema completo |
| Velocidad de creación | 5-30 segundos | 1-5 minutos |
| Uso de disco | 50-500 MB | 500 MB - 5 GB |
| Uso de RAM | Mínimo | Moderado-Alto |
| Curva de aprendizaje | Baja | Media-Alta |
| Portabilidad | Limitada | Total |
| Jupyter nativo | Sí | Complejo |
| Múltiples lenguajes | No | Sí |
| Bases de datos | Instalación aparte | Incluidas en contenedores |
| Estándar de producción | No recomendado | Sí |

## 🧭 Guía de decisión

### Usá entorno virtual cuando:

- Solo necesitás Python y sus librerías
- Estás aprendiendo o prototipando rápidamente
- Trabajás con notebooks de análisis
- Es un proyecto individual pequeño

### Usá Docker cuando:

- Necesitás múltiples servicios (base de datos, cache, etc.)
- Trabajás en equipo y necesitás consistencia
- Vas a desplegar a producción o a un VPS
- Querés garantizar reproducibilidad total

## ⚡ Estrategia híbrida — lo mejor de los dos mundos

### Analogía

Un cocinero profesional tiene su cuchillo personal (entorno virtual: rápido, conocido, siempre a mano para tareas concretas) y la cocina industrial del restaurante (Docker: con todos los equipos necesarios para el servicio completo). No elige uno u otro para siempre: usa el cuchillo para cortar y la cocina para servir.

### Dónde vive esto en el mundo real

Este patrón es el estándar en equipos profesionales: entorno virtual para el ciclo de desarrollo (editar, probar, iterar rápido) y Docker para el despliegue (reproducible, estable, compartible). Las herramientas de DevOps como GitHub Actions construyen contenedores Docker a partir del código que se desarrolló en entornos virtuales.

---

### El flujo profesional

```bash
# Fase 1 — Desarrollo (entorno virtual, rápido para iterar):
uv venv .venv
source .venv/bin/activate
uv pip install -r requirements.txt
# → editar, probar, corregir

# Fase 2 — Empaquetado (Docker, listo para distribuir):
docker build -t mi-app .

# Fase 3 — Despliegue (Docker en el servidor):
docker run --env-file .env -p 7860:7860 mi-app
```
