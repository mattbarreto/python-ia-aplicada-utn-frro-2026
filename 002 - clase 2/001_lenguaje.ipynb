## ◈ Microglosario

Antes de escribir la primera línea de código, alineamos el vocabulario.

| Término | Qué es en lenguaje llano |
|---|---|
| **Tipo de dato** | La categoría de un valor: número, texto, verdadero/falso. Python necesita saberlo para operar correctamente. |
| **Variable** | Un nombre que apunta a un valor en memoria. Puede cambiar mientras el programa corre. |
| **Condicional** | Una instrucción que elige un camino u otro según si una condición es verdadera o falsa. |
| **Iteración** | Repetir un bloque de código para cada elemento de una colección. |

| **Retorno** | El valor que una función entrega a quien la llamó. |

## ⚙️ Tipos de datos

### Analogía

Los tipos de datos son como los contenedores en una ferretería: un frasco para líquidos, una bolsa para polvos, una caja para piezas. Cada contenedor impone reglas sobre lo que puede guardar y cómo se opera con eso. Python necesita saber en qué tipo de contenedor está cada valor para poder sumar, comparar, transformar o mostrar.

### Dónde vive esto en el mundo real

Todo sistema de datos trabaja con tipos. Cuando un hospital registra una consulta, cada campo tiene un tipo asignado: el nombre del paciente es texto, la temperatura es decimal, el número de sala es entero, y si fue atendida o no es verdadero/falso. Esa distinción no es burocracia: determina cómo el sistema almacena, valida, procesa y muestra cada dato. Un error de tipo es uno de los bugs más comunes en producción.

En Python, los cuatro tipos básicos son:

| Tipo | Nombre en Python | Ejemplo |
|---|---|---|
| Número entero | `int` | `34`, `-7`, `0` |
| Número decimal | `float` | `87.5`, `3.14`, `-0.5` |
| Texto | `str` | `"Laura"`, `"activo"` |
| Verdadero / Falso | `bool` | `True`, `False` |

```python
# Registramos los datos de una participante del curso
nombre   = "Laura Méndez"    # texto: str
edad     = 34                 # número entero: int
puntaje  = 87.5               # número decimal: float
aprobada = True               # verdadero / falso: bool

print("✓ Datos registrados.")
```

```python
# type() nos dice exactamente qué tipo contiene cada variable
tipo_nombre   = type(nombre)
tipo_edad     = type(edad)
tipo_puntaje  = type(puntaje)
tipo_aprobada = type(aprobada)

print(f"nombre    → {tipo_nombre}")
print(f"edad      → {tipo_edad}")
print(f"puntaje   → {tipo_puntaje}")
print(f"aprobada  → {tipo_aprobada}")
```

```python
# Los formularios web siempre devuelven texto, aunque el campo sea numérico
# Esto es un caso muy frecuente en sistemas reales
edad_como_texto = "34"

# Convertimos el texto a entero para poder operar matemáticamente
edad_como_numero = int(edad_como_texto)

# Ahora sí podemos calcular
edad_en_diez_anios = edad_como_numero + 10

print(f"Tipo original  : {type(edad_como_texto)}")
print(f"Tipo convertido: {type(edad_como_numero)}")
print(f"En diez años   : {edad_en_diez_anios}")
```

## 🧭 Estructuras de control

### Analogía

El código sin control de flujo va siempre en línea recta, sin detenerse ni ramificarse. Las estructuras de control son las señales de tránsito: el semáforo decide si seguís o parás, el cartel de desvío te manda por otro camino, y la instrucción de dar la vuelta en manzana te hace repetir el recorrido hasta llegar al destino.

### Dónde vive esto en el mundo real

Toda lógica de negocio está construida con estructuras de control. Los sistemas de aprobación de crédito bancario son cadenas de `if` anidados. Los feeds de redes sociales son iteraciones sobre millones de publicaciones. Un sistema de alertas de temperatura en una planta industrial corre un `while` que verifica sensores cada pocos segundos. En este curso, van a usar estructuras de control en el script integrador de hoy, en el backend del Encuentro 4 y en los agentes del Encuentro 9.

### Condicionales — `if` / `elif` / `else`

Un condicional evalúa una expresión y ejecuta un bloque de código solo si esa expresión es verdadera. Python usa la **indentación** (4 espacios) para delimitar cada bloque.

```python
# Decidimos si Laura puede avanzar al siguiente módulo
puntaje_minimo = 70

if puntaje >= puntaje_minimo:
    print("✓ Puntaje suficiente. La participante avanza al siguiente módulo.")
else:
    print("✗ Puntaje insuficiente. Se recomienda refuerzo antes de avanzar.")
```

```python
# Con elif podemos manejar más de dos casos posibles
if puntaje >= 90:
    categoria = "Excelente"
elif puntaje >= 70:
    categoria = "Aprobada/o"
elif puntaje >= 50:
    categoria = "En proceso"
else:
    categoria = "Requiere refuerzo"

print(f"Participante : {nombre}")
print(f"Puntaje      : {puntaje}")
print(f"Categoría    : {categoria}")
```

### Iteraciones — `for`

El bucle `for` recorre una colección elemento por elemento y ejecuta el mismo bloque para cada uno. Es la herramienta principal para procesar listas, registros, respuestas de API y cualquier dato en volumen.

```python
# Lista de participantes registrados en el encuentro
participantes = ["Laura Méndez", "Sergio Vargas", "Marta Ríos", "Diego Paz", "Camila Torres"]

# Saludamos a cada participante por su nombre
for nombre_participante in participantes:
    print(f"Bienvenida/o, {nombre_participante}.")
```

```python
# Puntajes en el mismo orden que la lista de participantes
puntajes = [87.5, 63.0, 91.0, 48.5, 75.0]

# Recorremos ambas listas usando un índice numérico
for indice in range(len(participantes)):
    nombre_actual  = participantes[indice]   # nombre en la posición actual
    puntaje_actual = puntajes[indice]         # puntaje en la misma posición

    # Clasificamos según el puntaje
    if puntaje_actual >= 70:
        estado = "Aprobada/o"
    else:
        estado = "Refuerzo recomendado"

    print(f"{nombre_actual:20} | Puntaje: {puntaje_actual:5.1f} | {estado}")
```

## 🔧 Funciones

### Analogía

Una función es una receta de cocina: tiene un nombre, una lista de ingredientes (parámetros), un procedimiento y un resultado. Una vez que la receta existe, podés usarla cien veces sin volver a escribirla. La diferencia con una receta real es que en programación podés cambiar los ingredientes cada vez que la llamás — y siempre obtenés el resultado correcto para esos ingredientes.

### Dónde vive esto en el mundo real

Las funciones son la unidad básica de todo sistema de software. Cada endpoint de una API es una función expuesta al mundo. Cada botón en una app ejecuta al menos una. Los modelos de machine learning son, internamente, composiciones de miles de funciones matemáticas. En este curso, las funciones que escriban hoy van a ser los bloques que compongan el backend del Encuentro 4 y los agentes del Encuentro 9. Sin funciones, el código no escala ni se puede mantener.

```python
# Definimos una función: en este momento solo existe, todavía no hace nada
def saludar_participante(nombre_persona):
    saludo = f"Bienvenida/o al Encuentro 2, {nombre_persona}."
    return saludo

print("✓ Función definida. Existe, pero no se ejecutó todavía.")
```

```python
# Llamamos a la función — ahora sí se ejecuta
mensaje_laura  = saludar_participante("Laura Méndez")
mensaje_sergio = saludar_participante("Sergio Vargas")

print(mensaje_laura)
print(mensaje_sergio)
```

```python
# Una función más completa: recibe nombre y puntaje, devuelve un informe
def evaluar_participante(nombre_persona, puntaje_obtenido):
    # Determinamos la categoría según el puntaje
    if puntaje_obtenido >= 90:
        categoria_obtenida = "Excelente"
    elif puntaje_obtenido >= 70:
        categoria_obtenida = "Aprobada/o"
    else:
        categoria_obtenida = "Refuerzo recomendado"

    # Armamos el informe y lo devolvemos
    informe = f"{nombre_persona}: {puntaje_obtenido} pts → {categoria_obtenida}"
    return informe

print("✓ Función definida.")
```

```python
# Aplicamos la función a todos los participantes usando el for del bloque anterior
for indice in range(len(participantes)):
    nombre_actual  = participantes[indice]
    puntaje_actual = puntajes[indice]

    # La función encapsula toda la lógica: si cambia el criterio, lo cambiamos en un solo lugar
    resultado = evaluar_participante(nombre_actual, puntaje_actual)
    print(resultado)
```
