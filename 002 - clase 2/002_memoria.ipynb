## ◈ Microglosario

| Término | Qué es en lenguaje llano |
|---|---|
| **Archivo** | Un conjunto de datos guardado en disco con un nombre. Existe antes y después de que el programa corra. |
| **Modo de apertura** | La intención con la que abrimos un archivo: solo leer, escribir desde cero, o agregar al final. |
| **Context manager** | El bloque `with open()` que garantiza que el archivo se cierra correctamente, pase lo que pase. |
| **JSON** | Un formato de texto para representar datos estructurados: diccionarios, listas, números, texto y booleanos. |
| **Serialización** | Convertir un objeto de Python (dict, lista) a texto JSON para guardarlo o enviarlo. |
| **Excepción** | Un error que ocurre mientras el programa corre. Con `try/except` podemos anticiparlos y responder sin que el programa se detenga. |

## 🗒️ Archivos de texto

### Analogía

La memoria RAM de una computadora es como la mesa de trabajo: todo lo que está ahí desaparece cuando apagás. Un archivo en disco es el cajón: lo que guardás ahí sigue estando mañana, la semana que viene, o cuando otra persona abre la misma computadora. El programa de hoy puede dejarle datos al programa de mañana.

### Dónde vive esto en el mundo real

Los archivos son la forma más universal de intercambiar datos entre sistemas. Los logs de un servidor web son archivos de texto que un sistema de monitoreo lee para detectar errores. Los reportes de una campaña de salud se exportan como `.txt` o `.csv` para distribuirlos por mail. Los modelos de machine learning leen datasets desde archivos antes de entrenarse. Toda la cadena de datos de este curso — desde el scraping hasta el deploy — pasa en algún punto por la lectura y escritura de archivos.

---

Los tres modos de apertura más comunes:

| Modo | Significado | ¿Qué hace si el archivo ya existe? |
|---|---|---|
| `'r'` | Solo lectura | Lo abre tal cual |
| `'w'` | Escritura desde cero | Borra el contenido anterior |
| `'a'` | Agregar al final | Mantiene lo que había y suma al final |

```python
# Creamos un archivo con el registro inicial de la campaña de salud
# El bloque with garantiza que el archivo se cierra aunque haya un error
with open("registro_campaña.txt", "w", encoding="utf-8") as archivo:
    archivo.write("Campaña de Salud Comunitaria — Barrio Sur\n")
    archivo.write("Fecha de inicio: 2026-06-03\n")
    archivo.write("Responsable: Dra. Verónica Salinas\n")

print("✓ Archivo creado: registro_campaña.txt")
```

```python
# Leemos el archivo completo de una vez
with open("registro_campaña.txt", "r", encoding="utf-8") as archivo:
    contenido = archivo.read()

print("── Contenido del archivo ──────────────")
print(contenido)
```

```python
# Con 'a' agregamos una línea sin borrar lo que había
with open("registro_campaña.txt", "a", encoding="utf-8") as archivo:
    archivo.write("Total de beneficiarias/os atendidos: 47\n")

print("✓ Línea agregada al archivo.")
```

```python
# readlines() devuelve cada línea como un elemento de una lista
# Útil cuando queremos procesar el archivo línea por línea
with open("registro_campaña.txt", "r", encoding="utf-8") as archivo:
    lineas = archivo.readlines()

print(f"✓ El archivo tiene {len(lineas)} líneas.")
print()

# Mostramos cada línea numerada y sin el salto de línea final
for numero in range(len(lineas)):
    linea_limpia = lineas[numero].strip()    # strip() elimina espacios y saltos de línea
    print(f"  {numero + 1}. {linea_limpia}")
```

## 🧬 JSON — datos estructurados

### Analogía

Un archivo de texto plano es como una lista de compras escrita en papel: fácil de leer para una persona, difícil de procesar para un programa. JSON es como una planilla con filas y columnas claramente etiquetadas: cada campo tiene nombre y tipo, y cualquier sistema puede leerla e interpretarla sin ambigüedad.

### Dónde vive esto en el mundo real

JSON es el idioma de Internet. Cuando una app de clima muestra la temperatura de mañana, recibió esos datos como un JSON de una API. Cuando comprás en un e-commerce y el pago es aprobado, el banco devuelve un JSON con el resultado. Las bases de datos modernas como MongoDB guardan sus registros como documentos JSON. En este curso, vamos a consumir JSON desde APIs en el Bloque 3, recibirlo como salida de modelos de lenguaje en el Encuentro 6, y construir endpoints que lo devuelvan en el Encuentro 4.

```python
import json

print("✓ Módulo json cargado.")
```

```python
# Representamos a una beneficiaria de la campaña como un diccionario Python
beneficiaria = {
    "id": "B001",
    "nombre": "Ana Gutiérrez",
    "edad": 52,
    "presion_sistolica": 140,
    "glucosa": 98,
    "en_seguimiento": True
}

print("✓ Diccionario creado.")
print(f"Nombre: {beneficiaria['nombre']}")
print(f"Edad  : {beneficiaria['edad']}")
```

```python
# json.dumps() convierte el diccionario en un texto JSON (serialización)
# indent=2 lo formatea con sangría para que sea legible
texto_json = json.dumps(beneficiaria, ensure_ascii=False, indent=2)

print("── JSON resultante ─────────────────────")
print(texto_json)
```

```python
# json.loads() hace el camino inverso: texto JSON → diccionario Python
# Esto es lo que ocurre cuando recibimos una respuesta de una API
datos_recuperados = json.loads(texto_json)

print(f"Tipo del objeto recuperado : {type(datos_recuperados)}")
print(f"En seguimiento             : {datos_recuperados['en_seguimiento']}")
```

```python
# Lista de registros de la campaña — estructura más compleja
registros_campaña = [
    {"id": "B001", "nombre": "Ana Gutiérrez",   "edad": 52, "glucosa": 98,  "en_seguimiento": True},
    {"id": "B002", "nombre": "Carlos Ibáñez",   "edad": 67, "glucosa": 145, "en_seguimiento": True},
    {"id": "B003", "nombre": "Silvia Morales",  "edad": 44, "glucosa": 88,  "en_seguimiento": False},
    {"id": "B004", "nombre": "Roberto Leiva",   "edad": 71, "glucosa": 182, "en_seguimiento": True},
    {"id": "B005", "nombre": "Beatriz Campos",  "edad": 38, "glucosa": 79,  "en_seguimiento": False}
]

# Guardamos la lista en un archivo JSON
with open("registros_campaña.json", "w", encoding="utf-8") as archivo:
    json.dump(registros_campaña, archivo, ensure_ascii=False, indent=2)

print(f"✓ Guardados {len(registros_campaña)} registros en registros_campaña.json")
```

```python
# Cargamos el archivo JSON que acabamos de guardar
with open("registros_campaña.json", "r", encoding="utf-8") as archivo:
    datos_cargados = json.load(archivo)

print(f"✓ {len(datos_cargados)} registros cargados desde el archivo.")
print()

# Recorremos y mostramos solo los que están en seguimiento
print("── Beneficiarias/os en seguimiento activo ──")
for registro in datos_cargados:
    if registro["en_seguimiento"]:
        glucosa_actual = registro["glucosa"]
        nombre_actual  = registro["nombre"]
        print(f"  {nombre_actual:20} | Glucosa: {glucosa_actual} mg/dL")
```

## 🧭 Excepciones

### Analogía

Un programa sin manejo de errores es como un médico que solo sabe atender pacientes sanos. El mundo real genera situaciones inesperadas: el archivo no existe, el JSON está mal formado, la conexión se cortó. Las excepciones son el sistema de guardia: cuando algo sale mal, el programa no se derrumba — toma el camino alternativo que nosotros diseñamos.

### Dónde vive esto en el mundo real

Los sistemas productivos están llenos de `try/except`. Los backends de FastAPI capturan excepciones y devuelven respuestas HTTP con el código de error correcto en lugar de caerse. Los scripts de scraping manejan páginas que no cargan. Los pipelines de datos continúan procesando aunque un registro esté corrupto. En este curso, van a necesitar excepciones desde el Bloque 3 (APIs que pueden fallar) hasta el backend del Encuentro 4.

```python
# Sin manejo de errores: si el archivo no existe, el programa se cae
# Probamos primero el caso que funciona
try:
    with open("registros_campaña.json", "r", encoding="utf-8") as archivo:
        datos = json.load(archivo)
    print(f"✓ Archivo cargado. {len(datos)} registros encontrados.")

except FileNotFoundError:
    # Este bloque corre solo si el archivo no existe
    print("✗ El archivo no se encontró. Verificá la ruta y el nombre.")
```

```python
# Probamos el caso que falla: pedimos un archivo que no existe
try:
    with open("archivo_inexistente.json", "r", encoding="utf-8") as archivo:
        datos = json.load(archivo)
    print(f"✓ Archivo cargado.")

except FileNotFoundError:
    print("✗ El archivo no se encontró. Verificá la ruta y el nombre.")
```

```python
# Podemos capturar múltiples tipos de error en el mismo bloque try
nombre_archivo = "registros_campaña.json"

try:
    with open(nombre_archivo, "r", encoding="utf-8") as archivo:
        datos = json.load(archivo)          # puede fallar si el JSON está malformado
    print(f"✓ {len(datos)} registros cargados correctamente.")

except FileNotFoundError:
    # El archivo no existe en esa ruta
    print(f"✗ No se encontró el archivo '{nombre_archivo}'.")

except json.JSONDecodeError as error:
    # El archivo existe pero su contenido no es JSON válido
    print(f"✗ El archivo tiene formato inválido: {error}")

except Exception as error:
    # Cualquier otro error inesperado
    print(f"✗ Error inesperado: {error}")
```

```python
# finally corre siempre, haya error o no
# Útil para acciones de cierre, logs o limpieza
try:
    with open("registros_campaña.json", "r", encoding="utf-8") as archivo:
        datos = json.load(archivo)
    print(f"✓ Carga exitosa. {len(datos)} registros.")

except FileNotFoundError:
    print("✗ Archivo no encontrado.")

finally:
    # Este bloque siempre se ejecuta
    print("── Intento de carga finalizado. ─────")
```
